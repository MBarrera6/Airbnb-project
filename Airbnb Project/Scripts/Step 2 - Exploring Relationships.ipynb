```python
df2['pricing'].value_counts(dropna=True)
```

```python
# creating a categorical plot crossing the price with neighbourhood_group
sns.set(style='ticks')
g = sns.catplot(x= 'neighbourhood_group', y='price', hue = 'pricing', data = df2)
```

```python
# creating listing price categories
df2.loc[df2['availability_365'] <100, 'popularity'] = 'popular'
df2.loc[df2['availability_365'] >200, 'popularity'] = 'non_popular'
df2.loc[(df2['availability_365'] >=100) & (df2['availability_365'] <=200), 'popularity'] = 'normal'
```
